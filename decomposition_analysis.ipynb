以Taylor展開方法分析運具排放清冊各變數影響公路溫室氣體排放效果
==============================================================
本研究在於探討公路運輸部門之溫室氣體排放量歷年的波動(Variations)來源為何，利用 Taylor展開方法的方式拆解各個變數的波動程度對應於公路部門總合之波動程度，如此便可以再更進一步細看各個變數各年對於溫室氣體排放的影響大小，例如在2007、2008年的時候可以觀察到這兩年的公路運輸溫室氣體排放呈現下降的情形，而究竟是哪一個變數的波動為主要造成下降之原因，或是此變數的變化幅度又如何影響整體的排放量，是無法利用傳統的計量方法討論的。

為了估算我國運輸部門歷年之溫室氣體排放量，交通部運輸研究所會每年更新運輸部門運具排放清冊，其中將運輸部門之溫室氣體排放來源依照運輸型態分為以下4種來源：
1. 公路運輸：舉凡所有公路運輸方式，依運具別分類有小客車、小貨車、大客車、大貨車與機車等。
2. 軌道運輸：鐵路、高鐵、捷運	
3. 國內水路運輸：
4. 國內航空運輸： 我國溫室氣體減量及管理法所規範之範圍僅限於國內航空部份，國際航空並不屬於溫管法所屬，因此排放清冊只有估算國內航空部份。

而公路運輸的溫室氣體排放量即佔整體運輸部門排放量近96%，考量到處理不同運輸型態的會增加分析的複雜度，因此現階段本研究僅以公路部門作為研究主題。

根據交通部運輸研究所出版之[城際運輸節能減碳策略評估模組開發及應用](https://w3.iot.gov.tw/Modules/Publication/Publication-Details?node=3598166a-c839-4f0d-8beb-33c093e49de0&id=64106ecf-6856-4a12-8ac0-d146dbf3f2f9)，公路運輸的溫室氣排放量的計算方式如下：

$$公路溫室氣體排放量_{ij} = 公路車輛登記數_{ij} \times 公路每車年行駛里程_{ij} \div 公路燃油效率_{ij} \times 能源平衡表調整係數_{j} \times 燃料溫室氣體排放係數_{j} \div 1000,000$$
- $i$ 為車種別，包括自用小客車、營業小客車、自用小貨車、營業小貨車、自用大客車、遊覽車、公車與客運車、自用大貨車、營業大貨車、機車等車種； 

- $j$ 為燃料別，包括汽油、柴油、LPG；

在考量不同車種、不同使用方式以及不同的燃料別之後，此每一年的影響溫室氣體排放量的變數一共有51個，並且各個變數每年的變動程度和方向皆不同。本研究將會分為 個架構進行：
1. 資料匯入

3. 分析結果繪圖
4. 結論

# Taylor展開方法分析
本研究所使用的方法主要為多變數Taylor展開式將分解各個變數對溫室氣體排放的影響，

多變數Taylor展開式的公式如下：
$$ F(X_1, \cdots ,X_n) \approx F(\tilde{X}_1, \cdots ,\tilde{X}_n) + \sum^{n}_{j=1}{F_{X_j}(\tilde{X}_1, \cdots ,\tilde{X}_n)(X_j - \tilde{X}_j)} + \sum^{n}_{j=1}{\sum^{n}_{K=1}}{F_{X_{j} X_k}(\tilde{X_1}, \cdots ,\tilde{X_n}) \cdot (X_j - \tilde{X}_j)(X_k - \tilde{X}_k)} + \cdots $$

- $j,k$：各運具分類，如汽油自用小客車、柴用營業大貨車

- $n$：運具種類個數

- $X$：各項變數值，如公路車輛登記數、公路燃油效率

- $\tilde{X}$：前一期變數值

- $F(X_1, \cdots ,X_n)$：設為本期公路溫室氣體排放量函數

- $F(\tilde{X}_1, \cdots ,\tilde{X}_n)$：設為上期公路溫室氣體排放量函數

- $F(X_1, \cdots ,X_n) - F(\tilde{X}_1, \cdots ,\tilde{X}_n)$：對溫室氣體排放量進行差分，得出的值即為溫室氣體變化量

- $F_{X_j}(\tilde{X}_1, \cdots ,\tilde{X}_n)$：公路溫室氣體排放量函數對$X_j$之一階偏微分

- $X_j - \tilde{X}_j$：對各變數進行差分，即為各變數之變化量

- $\sum^{n}_{j=1}{F_{X_j}(\tilde{X}_1, \cdots ,\tilde{X}_n)(X_j - \tilde{X}_j)}$：第1項Taylor展開

- $\sum^{n}_{j=1}{\sum^{n}_{K=1}}{F_{X_{j} X_k}(\tilde{X_1}, \cdots ,\tilde{X_n}) \cdot (X_j - \tilde{X}_j)(X_k - \tilde{X}_k)}$：第2項Taylor展開

將上期溫室氣體排放量$F(\tilde{X}_1, \cdots ,\tilde{X}_n)$移至左手邊，並將右手邊的Taylor展開式取2項整理之後可得：

$$ F(X_1, \cdots ,X_n) - F(\tilde{X}_1, \cdots ,\tilde{X}_n) \approx \sum^{n}_{j=1} \Big[ {F_{X_j}(\tilde{X}_1, \cdots ,\tilde{X}_n)(X_j - \tilde{X}_j)} + {\sum^{n}_{K=1}}{F_{X_{j} X_k}(\tilde{X_1}, \cdots ,\tilde{X_n}) \cdot (X_j - \tilde{X}_j)(X_k - \tilde{X}_k)} \Big]$$

計算右手邊的$ \ j=1 \cdots n$ 即可得到拆解後各變數的變化量對於溫室氣體排放的變化量的影響值

例如 $ \ j=1 $ 時，即可得到$X_1$變數的影響值：

$$ \Delta GHG_{X_1} = X_1之變化影響總體溫室氣體排放量 =  {F_{X_1}(\tilde{X}_1, \cdots ,\tilde{X}_n)(X_1 - \tilde{X}_1)} + {\sum^{n}_{K=1}}{F_{X_{1} X_k}(\tilde{X_1}, \cdots ,\tilde{X_n}) \cdot (X_1 - \tilde{X}_1)(X_k - \tilde{X}_k)} $$

假設$X_1$為2007年汽油自用小客車的數量，用此方法便可以拆解出「因為2007年汽油自用小客車的數量的變化所造成之溫室氣體排放量」

## 資料匯入函式

```python
# 匯入運具排放清冊資料
def import_emission_data(sheet_name,skiprows, skip_footer, usecols, dropcol, columns=vehicles,
                         filename="1990_2015年運輸部門運具別排放清冊1060203.xlsx", index_col=0, na_values="-", convert_float=True):
    '''匯入運具排放清冊資料
    主要是使用pandas.read_excel匯入，但是每份資料格式皆有些出入，因此要使用skiprows, skip_footer, usecols, dropcol等參數進行微調讀取 
    '''
    df = pd.read_excel(filename, sheet_name=sheet_name, index_col=index_col,
                       skiprows=skiprows, skip_footer=skip_footer, usecols=usecols, na_values=na_values, convert_float=True)
    df.index.name = 'year'
    df.drop(columns=df.columns[dropcol], inplace=True)
    if len(columns)!=0:
        df.columns = columns
    return df
```

因為原先的排放清冊中有計算完之後的總量，為了檢查我們的計算方式是否正確，這裡建立一個檢查的函式

```python
# 比較計算結果與排放清冊差異
def compare(calculateddata, basedata):
    '''比較計算結果與排放清冊差異
    利用 計算/匯入 之後只要顯示有差異的值，無差異值則顯示NA 
    '''
    errorterm = calculateddata / basedata
    errorterm = errorterm[(errorterm.notna()) & (errorterm!=1)]
    return errorterm
```

# 二、使用Taylor 展開分析演算法
本研究所使用的方法主要為多變數Taylor展開式將分解各個變數對溫室氣體排放的影響，

多變數Taylor展開式的公式如下：
$$ F(X_1, \cdots ,X_n) \approx F(\tilde{X}_1, \cdots ,\tilde{X}_n) + \sum^{n}_{j=1}{F_{X_j}(\tilde{X}_1, \cdots ,\tilde{X}_n)(X_j - \tilde{X}_j)} + \sum^{n}_{j=1}{\sum^{n}_{K=1}}{F_{X_{j} X_k}(\tilde{X_1}, \cdots ,\tilde{X_n}) \cdot (X_j - \tilde{X}_j)(X_k - \tilde{X}_k)} + \cdots $$

- $X_j$：各個變數值

- $j,k$：各運具分類

- $F(X_1, \cdots ,X_n)$：設為本期溫室氣體排放量

- $F(\tilde{X}_1, \cdots ,\tilde{X}_n)$：設為上期溫室氣體排放量

- $F(X_1, \cdots ,X_n) - F(\tilde{X}_1, \cdots ,\tilde{X}_n)$：對溫室氣體排放量進行差分，得出的值即為溫室氣體變化量

- $F_{X_j}(\tilde{X}_1, \cdots ,\tilde{X}_n)$：一階微分

- $X_j - \tilde{X}_j$：對各變數進行差分，即為各變數之變化量

- $\sum^{n}_{j=1}{F_{X_j}(\tilde{X}_1, \cdots ,\tilde{X}_n)(X_j - \tilde{X}_j)}$：第1項Taylor展開

- $\sum^{n}_{j=1}{\sum^{n}_{K=1}}{F_{X_{j} X_k}(\tilde{X_1}, \cdots ,\tilde{X_n}) \cdot (X_j - \tilde{X}_j)(X_k - \tilde{X}_k)}$：第2項Taylor展開

將上期溫室氣體排放量$F(\tilde{X}_1, \cdots ,\tilde{X}_n)$移至左手邊，並將右邊整理之後可得：

$$ F(X_1, \cdots ,X_n) - F(\tilde{X}_1, \cdots ,\tilde{X}_n) \approx \sum^{n}_{j=1} \Big[ {F_{X_j}(\tilde{X}_1, \cdots ,\tilde{X}_n)(X_j - \tilde{X}_j)} + {\sum^{n}_{K=1}}{F_{X_{j} X_k}(\tilde{X_1}, \cdots ,\tilde{X_n}) \cdot (X_j - \tilde{X}_j)(X_k - \tilde{X}_k)} \Big] + \cdots $$

計算右手邊的$ \ j=1 \cdots n$ 即可得到拆解後各變數的變化量對於溫室氣體排放的變化量的影響值

例如 $ \ j=1 $ 時，即可得到$X_1$變數的影響值：

$$X_1之變化影響總體溫室氣體排放量 =  {F_{X_1}(\tilde{X}_1, \cdots ,\tilde{X}_n)(X_1 - \tilde{X}_1)} + {\sum^{n}_{K=1}}{F_{X_{1} X_k}(\tilde{X_1}, \cdots ,\tilde{X_n}) \cdot (X_1 - \tilde{X}_1)(X_k - \tilde{X}_k)} $$

## 資料分析初始化設定

## Taylor資料分析函式
### 用sympy.lambdify就可以把sympy計算出來的方程式用dataframe資料帶入
 - sympyFunc = lambdify((var1,var2,var3),func)
 - sympyFunc(df1,df2,df3)

```python
def dataframedify(symfunc):
    '''
    # 把sympy的計算出的函數結果中的變數(num,avgdst,fueleff...)用dataframe的變數代入(road_veh_num,road_veh_avgdst...),
    之後會return 一個dataframe
    '''
    f = lambdify(tuple(symdict.values()), symfunc)
    return f(*dfdict.values())
```

$$f_x dx =  \text{diff(f,x)} \times \text{x.diff()}$$
$\text{diff(f,x)：} f_x $

$\text{x.diff()：} dx $

```python
def first_taylor_term(symfunc, key):
    '''取Taylor展開的微分一次式計算
    輸入函式以及分析目標變數後回傳該目標變數的Taylor一階分解Dataframe
    '''
    fx = diff(symfunc,symdict[key])
    output = dataframedify(fx) * dfdict[key].diff()
    return output
```

$$
- \frac{1}{2!} \big( f_{xx} dx dx + f_{xy} dx dy +f_{xz} dx dz +f_{xw} dx dw +f_{xt} dx dt \big)
$$

```python
def second_taylor_term(symfunc,key):
    '''取Taylor展開的微分二次式計算
    輸入函式以及分析目標變數後回傳該目標變數的Taylor二階分解Dataframe
    '''
    fx = diff(symfunc,symdict[key])
    output = pd.DataFrame()
    for k,s in symdict.items():
        fxx = diff(fx,s)
        output = output.add( - 1/2 *dataframedify(fxx)*dfdict[key].diff()*dfdict[k].diff(),fill_value=0) # fxx*dx*dx
    return output
```

```python
def taylor_decompose(symfunc,key):
    '''
    將一階與二階的分解結果加總，日後可以擴充增加三階
    '''
    return first_taylor_term(symfunc,key) + second_taylor_term(symfunc,key)
```

```python
# (檢查!)比較只取 first order 和取到 second order 與原始值的差異
testtotal = pd.DataFrame()
for k in dfdict.keys():
    testtotal = testtotal.add(first_taylor_term(GHGemit,k),fill_value=0)
firsttotal = compare(testtotal,road_veh_GHGemit.diff()).mean(axis=1) # 這裡是用平均值計算，平均誤差為多少

testtotal = pd.DataFrame()
for k in dfdict.keys():
    testtotal = testtotal.add(taylor_decompose(GHGemit,k), fill_value=0) 
secondtotal = compare(testtotal,road_veh_GHGemit.diff()).mean(axis=1) # 這裡是用平均值計算，平均誤差為多少
    
first_second_compare = pd.concat([firsttotal,secondtotal],axis=1)
first_second_compare.columns = ['取1項','取2項']
print('比較Taylor展開方法只取1項和取2項的拆解結果加總後與原始值之差異，愈接近1愈好。')
first_second_compare.tail()
# 取2項的誤差值比只有取1項的值來得小
```

經過Taylor展開模型分解後的資料的平均誤差，只取1項的平均誤差大約是在15%左右，取2項之後的平均誤差則不到1%

```python
def most_influence_of(df,year,keep=3):
    '''取出DataFrame中該年影響力最大的正向影響變數以及負向影響變數
    df:DataFrame    
    year:int，目標年    
    keep:int，取出影響最大的變數個數
    '''
    assert year in year_range, '輸入的年份({0})並不在資料年份中({1}-{2})'.format(year,min(year_range),max(year_range))
    assert keep*2 < len(df.columns), '要取出的變數過多, keep必須小於%d' %(len(df.columns)//2)
    
    select_series = df.loc[year,:]
    nlarge = select_series.nlargest(keep).sort_values(ascending=False)
    nsmall = select_series.nsmallest(keep).sort_values(ascending=False)
    
    return pd.DataFrame(nlarge.append(nsmall))
```

```python
def stack(stacklist):
    '''輸入變數的list的之後，回傳一份可供plt.bar堆疊的Dataframe
    stacklist = ['petrol_balcoeff', 'diesel_balcoeff', 'car_pu_petrol_num', 'car_pu_petrol_avgdst']
    '''
    stack_layer = pd.DataFrame(index=year_range)
    bound = pd.DataFrame(0, index=year_range ,columns=['upper','lower','target'])

    for i, target in enumerate(stacklist):
        if i == 0:
            bound['target'] = taylor_result_print[target]
            stack_layer[target] = 0

            # 更新上下邊界
            bound['upper'] = bound.apply(lambda ser: ser['upper']+ser['target'] if ser['target']>0 else ser['upper'], axis=1)
            bound['lower'] = bound.apply(lambda ser: ser['lower']+ser['target'] if ser['target']<0 else ser['lower'], axis=1)
        else:
            bound['target'] = taylor_result_print[target]
            stack_layer[target] = bound.apply(lambda ser: ser['upper'] if ser['target']>0 else ser['lower'], axis=1)

            # 更新上下邊界
            bound['upper'] = bound.apply(lambda ser: ser['upper']+ser['target'] if ser['target']>0 else ser['upper'], axis=1)
            bound['lower'] = bound.apply(lambda ser: ser['lower']+ser['target'] if ser['target']<0 else ser['lower'], axis=1)
    return stack_layer
```
